```python
# Load CSV, detect columns, and plot train vs validation cross-entropy over epochs
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

csv_path = Path("/Users/rocioperez/Documents/Projects/AFG/Git_Group_Project/eeg-prism/data/training_history.csv")
df = pd.read_csv(csv_path)
```

```python
df.head()
```

```python
plt.figure(figsize=(6, 4))
plt.title("EEGNet Model Loss")
plt.plot(df['epoch'], df['train_loss'], label="train")
plt.xlabel("Epoch")
plt.ylabel("Categorical cross-entropy")
plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(6, 4))
plt.title("EEGNet Model Accuracy")
plt.plot(df['epoch'], df['train_accuracy'], label="train")
#plt.plot(df['epoch'], df['valid_accuracy'], label="validation")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(6, 4))
plt.title("EEGNet Model Loss")
plt.plot(df['epoch'], df['train_loss'], label="train")
plt.plot(df['epoch'], df['valid_loss'], label="validation")
plt.xlabel("Epoch")
plt.ylabel("Categorical cross-entropy")
plt.legend()
plt.tight_layout()
```
